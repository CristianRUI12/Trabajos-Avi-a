## Clasificación

```python
resultados = []

for comentario in comentarios:
    analisis = TextBlob(comentario)
    polaridad = analisis.sentiment.polarity

    if polaridad > 0:
        resultados.append("Positivo")
    elif polaridad < 0:
        resultados.append("Negativo")
    else:
        resultados.append("Neutro")
```

## Conteo

```python
positivos = resultados.count("Positivo")
negativos = resultados.count("Negativo")
neutros = resultados.count("Neutro")

print("Positivos:", positivos)
print("Negativos:", negativos)
print("Neutros:", neutros)
```

## Gráfica

```python
etiquetas = ["Positivos", "Negativos", "Neutros"]
valores = [positivos, negativos, neutros]

plt.bar(etiquetas, valores)
plt.title("Análisis de Sentimientos")
plt.show()
```
