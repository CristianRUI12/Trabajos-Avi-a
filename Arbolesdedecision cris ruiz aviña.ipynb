## Explicación del código

```python
from sklearn.tree import DecisionTreeClassifier

modelo = DecisionTreeClassifier(max_depth=50)

X_ent_d = pd.get_dummies(x_ent)
X_pru_d = pd.get_dummies(x_pru).reindex(columns=X_ent_d.columns, fill_value=0)

modelo.fit(X_ent_d, y_ent)
predicciones = modelo.predict(X_pru_d)
```

## Explicación del código
Este bloque ejecuta instrucciones relacionadas con el modelo de árbol de decisiones, incluyendo preparación de datos, entrenamiento o evaluación según corresponda.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

resultados = []
for i in range(1,15):
    modelo = DecisionTreeClassifier(max_depth=i)
    modelo.fit(X_ent_d, y_ent)
    predicciones = modelo.predict(X_pru_d)
    exactitud = accuracy_score(y_pru, predicciones)
    print(f"Resultado para {i}: {exactitud}")
    resultados.append(exactitud)
```
