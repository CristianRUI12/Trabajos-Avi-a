```python
"""
Traduce palabras usando un diccionario simple.
"""

dic = {
    "hola": "hello",
    "mundo": "world",
    "perro": "dog",
    "gato": "cat",
    "casa": "house",
    "agua": "water",
    "libro": "book",
    "amigo": "friend",
    "escuela": "school",
    "comida": "food"
}

def translate(word):
    return dic.get(word, "no encontrado")

# Traducir 10 palabras
words = ["hola", "mundo", "perro", "gato", "casa", "agua", "libro", "amigo", "escuela", "comida"]

for word in words:
    print(word, "→", translate(word))
```
